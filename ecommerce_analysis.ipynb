```python
import mysql.connector
import pandas as pd
```

```python
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="3030",
    database="ecommerce_analytics"
)
```

```python
connection.is_connected()
```

```python
#customers
```

```python
customers_df = pd.read_sql("SELECT * FROM customers", connection)
customers_df.head()
```

```python
#products
```

```python
products_df = pd.read_sql("SELECT * FROM products", connection)
products_df.head()
```

```python
#orders
```

```python
orders_df = pd.read_sql("SELECT * FROM orders", connection)
orders_df.head()
```

```python
#ordre_items
```

```python
order_items_df = pd.read_sql("SELECT * FROM order_items", connection)
order_items_df.head()
```

```python
#basic check
```

```python
customers_df.info()
```

```python
products_df.info()
```

```python
orders_df.info()
```

```python
order_items_df.info()
```

```python
#create dataset(combine all tables)
```

```python
#orders+customers
```

```python
df = pd.merge(orders_df, customers_df, on = "customer_id")
```

```python
#add order items
```

```python
df = pd.merge(df, order_items_df, on = "order_id")
```

```python
#add products
```

```python
df = pd.merge(df, products_df, on = "product_id")
```

```python
df.head()
```

```python
#total revenue
```

```python
df.columns
```

```python
df["total_revenue"] = df["quantity"] * df["price_x"]
```

```python
df.head()
```

```python
#analysis
```

```python
#total_revenue
```

```python
df["total_revenue"].sum()
```

```python
#top customers
```

```python
df.groupby("customer_name")["total_revenue"].sum().sort_values(ascending=False)
```

```python
#sales by city
```

```python
df.groupby("location")["total_revenue"].sum().sort_values(ascending=False)
```

```python
#best selling_products
```

```python
df.groupby("product_name")["total_revenue"].sum().sort_values(ascending=False)
```

```python
#category wise sales
```

```python
df.groupby("category")["total_revenue"].sum()
```

```python
#save output
```

```python
df.to_csv("final_dataset.csv",index=False)
```
